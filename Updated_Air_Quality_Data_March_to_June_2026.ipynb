```python
import pandas as pd
import requests
import time
import plotly.express as px
```

```python
cities = {
    "Lahore": {"lat": 31.5204, "lon": 74.3587},
    "Karachi": {"lat": 24.8607, "lon": 67.0011},
    "Islamabad": {"lat": 33.6844, "lon": 73.0479},
    "Rawalpindi": {"lat": 33.5651, "lon": 73.0169},
    "Faisalabad": {"lat": 31.4504, "lon": 73.1350},
    "Multan": {"lat": 30.1575, "lon": 71.5249},
    "Peshawar": {"lat": 34.0151, "lon": 71.5249},
    "Quetta": {"lat": 30.1798, "lon": 66.9750},
    "Rahim Yar Khan": {"lat": 28.4197, "lon": 70.3035},
    "Sialkot": {"lat": 32.4945, "lon": 74.5229}
}
```

```python
start_date = "2026-03-01"
end_date = "2026-06-26"   # aaj ki date tak; agar tum June 30 tak chaho to "2026-06-30" kar do

all_city_data = []

for city, coord in cities.items():
    print("Downloading data for:", city)

    url = "https://air-quality-api.open-meteo.com/v1/air-quality"

    params = {
        "latitude": coord["lat"],
        "longitude": coord["lon"],
        "hourly": "pm10,pm2_5,carbon_monoxide,nitrogen_dioxide,sulphur_dioxide,ozone,us_aqi,us_aqi_pm2_5,us_aqi_pm10",
        "start_date": start_date,
        "end_date": end_date,
        "timezone": "Asia/Karachi"
    }

    response = requests.get(url, params=params)

    print(city, "Status:", response.status_code)

    data = response.json()

    if "hourly" in data:
        temp_df = pd.DataFrame(data["hourly"])
        temp_df["city"] = city
        temp_df["latitude"] = coord["lat"]
        temp_df["longitude"] = coord["lon"]
        all_city_data.append(temp_df)
    else:
        print("Problem for", city)
        print(data)

    time.sleep(1)

updated_df = pd.concat(all_city_data, ignore_index=True)

updated_df.head()
```

```python
updated_df["time"] = pd.to_datetime(updated_df["time"])

updated_df["date"] = updated_df["time"].dt.date
updated_df["hour"] = updated_df["time"].dt.hour
updated_df["month"] = updated_df["time"].dt.strftime("%b %Y")

updated_df.head()
```

```python
print("Start Date:", updated_df["time"].min())
print("End Date:", updated_df["time"].max())

updated_df["month"].value_counts().sort_index()
```

```python
df = updated_df.copy()

city_col = "city"
date_col = "time"
pollutant_col = "pm2_5"
aqi_col = "us_aqi"

df.head()
```

```python
def classify_aqi(aqi):
    if pd.isna(aqi):
        return "Unknown"
    elif aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Moderate"
    elif aqi <= 150:
        return "Unhealthy for Sensitive Groups"
    elif aqi <= 200:
        return "Unhealthy"
    elif aqi <= 300:
        return "Very Unhealthy"
    else:
        return "Hazardous"

df["aqi_category"] = df[aqi_col].apply(classify_aqi)

risk_score_map = {
    "Good": 1,
    "Moderate": 2,
    "Unhealthy for Sensitive Groups": 3,
    "Unhealthy": 4,
    "Very Unhealthy": 5,
    "Hazardous": 6,
    "Unknown": 0
}

df["health_risk_score"] = df["aqi_category"].map(risk_score_map)

df[["city", "time", "pm2_5", "us_aqi", "aqi_category", "health_risk_score"]].head()
```

```python
df.to_csv("updated_pakistan_air_quality_mar_jun_2026.csv", index=False)

print("Updated dataset saved successfully!")
print(df.shape)
```

```python
monthly_trend = df.groupby(["month", "city"], as_index=False)["pm2_5"].mean()

month_order = ["Mar 2026", "Apr 2026", "May 2026", "Jun 2026"]

fig = px.line(
    monthly_trend,
    x="month",
    y="pm2_5",
    color="city",
    title="Updated Monthly PM2.5 Pollution Trend by City: March to June 2026",
    markers=True,
    category_orders={"month": month_order}
)

fig.show()
```

## Task 1: Web Scraping

In this section, I scraped air pollution and smog-related news headlines for Pakistan. The purpose is to understand how air pollution is being discussed in public/news sources.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

query = "Pakistan air pollution smog"
url = "https://news.google.com/rss/search?q=" + query.replace(" ", "+") + "&hl=en-PK&gl=PK&ceid=PK:en"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

print("Status Code:", response.status_code)

soup = BeautifulSoup(response.content, "xml")

items = soup.find_all("item")

news_data = []

for item in items[:50]:
    title = item.title.text if item.title else None
    link = item.link.text if item.link else None
    pub_date = item.pubDate.text if item.pubDate else None
    source = item.source.text if item.source else None

    news_data.append({
        "title": title,
        "source": source,
        "published_date": pub_date,
        "link": link
    })

news_df = pd.DataFrame(news_data)

display(news_df.head())
print("Total headlines scraped:", news_df.shape[0])
```

## Task 4: Sentiment Analysis

In this section, I applied sentiment analysis on the scraped news headlines. Each headline is classified as Positive, Neutral, or Negative.

```python
import nltk

nltk.download("vader_lexicon")

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):
    score = sia.polarity_scores(str(text))["compound"]

    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

news_df["sentiment"] = news_df["title"].apply(get_sentiment)

display(news_df.head(10))
```

```python
sentiment_counts = news_df["sentiment"].value_counts().reset_index()
sentiment_counts.columns = ["sentiment", "count"]

display(sentiment_counts)
```

```python
fig = px.bar(
    sentiment_counts,
    x="sentiment",
    y="count",
    title="Sentiment Analysis of Pakistan Air Pollution News Headlines",
    text="count"
)

fig.show()
```
